```python
#ANSWER 1: Time Series Analysis of Releases

#Dataset & Relevant Columns:

#Loading the Netflix dataset
netflix_data = pd.read_csv('netflix_titles.csv')

#Inspecting the relevant columns
print(netflix_data.columns)

#Counting the number of titles released per year
titles_per_year = netflix_data['release_year'].value_counts().sort_index()

#Analyzing the count of titles released per year
plt.figure(figsize=(12, 6))
titles_per_year.plot(kind='bar', color='skyblue')
plt.title('Count of Titles Released Per Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Visualization Tasks:

#Creating a line chart showing the yearly count of new titles
plt.figure(figsize=(12, 6))
plt.plot(titles_per_year.index, titles_per_year.values, marker='o', linestyle='-', color='blue')
plt.title('Yearly Count of New Titles on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid()
plt.show()

#Creating a bar chart showing the same yearly count
plt.figure(figsize=(12, 6))
titles_per_year.plot(kind='bar', color='skyblue')
plt.title('Yearly Count of New Titles on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Storytelling Task:

#Clarity and Readability:
#Line Chart: The line chart provides a continuous view of the data, making it easier to see the overall trend in the yearly count of new titles. 
#The smoothness of the line allows viewers to quickly grasp the direction of change over time.
#Bar Chart: The bar chart presents discrete values for each year, which can be visually impactful for comparing specific counts. 
#However, this requires more effort to interpret trends since the viewer must analyze the height of each bar individually.

#Trend Detection:
#Line Chart: Line chart excels in trend detection, as the connected points create a visual flow which highlights increase and decrease in title releases. 
#Viewers can easily spot upward or downward trends and observe the rate of change over the years.
#Bar Chart: While the bar chart can show changes in counts, it does not convey the same sense of continuity. 
#The viewer has to look at each bar separately to understand the overall trend, which can be less intuitive.

#Seasonal Patterns:
#Line Chart: If the data includes seasonal variations, the line chart can effectively illustrate these patterns over time. 
#The smooth transitions between points can highlight seasonal peaks, making it easier to identify recurring trends.
#Bar Chart: Seasonal patterns are not that clear in a bar chart, as the separation of bars can disturb the flow of data. 
#While it can show yearly totals, it may not effectively communicate seasonal fluctuations.


#Hence,for forecasting purpose, the line chart is the preferred choice. 
#Its ability to clearly show trends over time allows better predictions of future title releases based on historical data. 
#The continuous nature of the line chart helps in identifying patterns that can inform future content strategies, making it a more effective tool for forecasting than the bar chart.
```

```python
#ANSWER 2:Genre Distribution Analysis

#Dataset & Relevant Columns:

#Loading the Netflix dataset
netflix_data = pd.read_csv('netflix_titles.csv')

#Inspecting the relevant columns
print(netflix_data.columns)
genres_series = netflix_data['listed_in'].str.split(',').explode().str.strip()

#Counting the frequency of each genre
genre_counts = genres_series.value_counts()

#Picking the top genres
top_genres = genre_counts.head(10)

#Analyzing the frequency of top genres
plt.figure(figsize=(12, 6))
top_genres.plot(kind='bar', color='skyblue')
plt.title('Top Genres on Netflix')
plt.xlabel('Genres')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Visualization Tasks:

# Creating a bar chart to show frequency of the top genres
plt.figure(figsize=(12, 6))
top_genres.plot(kind='bar', color='skyblue')
plt.title('Top Genres on Netflix')
plt.xlabel('Genres')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()

# Creating a pie chart to show the genre distribution
plt.figure(figsize=(10, 10))
plt.pie(top_genres, labels=top_genres.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Genre Distribution on Netflix')
plt.axis('equal')  
plt.show()
```

```python
#Storytelling Task:

#Ease of Interpretation:
#Bar Chart: The bar chart provides a straightforward way to interpret the frequency of each genre. 
#The length of each bar directly tells with the number of titles in that genre, making it easy for viewers to quickly grasp which genres are most prevalent. 
#Pie Chart: The pie chart, presents the genre distribution as a whole, with each slice representing a genre's proportion of the total.
#While visually appealing, pie charts can sometimes be less intuitive, especially when there are many categories. 

#Precision in Comparing Values:
#Bar Chart: One of the significant advantages of the bar chart is its ability to facilitate precise comparisons between different genres.
#Viewers can easily compare the lengths of the bars side by side, allowing for quick assessments of which genres dominate and which are less represented. 
#This clarity is particularly beneficial when the differences in frequency are small, as the visual representation allows for more accurate comparisons.
#Pie Chart: On the other hand, the pie chart can make it challenging to compare values directly. 
#While it effectively shows the overall distribution, it is often difficult to determine the exact proportions of each genre without relying on the percentage labels. 
#This limitation can lead to misinterpretation, especially when slices are similar in size.

#Visual Appeal:
#Bar Chart: Bar charts are generally considered more versatile and can be designed to be visually striking with the use of colors, patterns, and labels. 
#They can accommodate a larger number of categories without becoming cluttered, making them suitable for datasets with many genres.
#Pie Chart: Pie charts are often used for their aesthetic appeal and ability to convey a sense of completeness. 
#They can be visually engaging, especially when using vibrant colors and clear labels. However, their effectiveness diminishes as the number of categories increases, leading to a cluttered and confusing visual.

#In this case, the bar chart is more appropriate for displaying genre distribution in the Netflix dataset. 
#The ease of interpretation, precision in comparing values, and overall clarity make it a better choice for analyzing the frequency of genres. 
#However, pie charts can be visually appealing and effective for showing proportions, they fail in providing the level of detail and accuracy needed for precise comparisons among multiple categories. 
#Hence, for a dataset with numerous genres, the bar chart is the best option as it is a more effective visualization tool.
```

```python
#ANSWER 3:Regional Content Analysis

#Dataset & Relevant Columns:

#Loading the Netflix dataset
netflix_data = pd.read_csv('netflix_titles.csv')

##Inspecting the relevant columns
print(netflix_data.columns)

#Counting the number of titles per country
country_counts = netflix_data['country'].value_counts()

#Analyzing the number of titles per primary country
country_counts = netflix_data['country'].value_counts()
top_10_countries = country_counts.head(10)
plt.figure(figsize=(12, 6))
top_10_countries.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Countries by Number of Titles on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Visualization Tasks:

#Creating a horizontal bar chart for country distribution
plt.figure(figsize=(10, 6))
country_counts[:10].plot(kind='barh', color='lightblue')
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.title("Top 10 Countries by Netflix Content (Bar Chart)")
plt.gca().invert_yaxis()  # Invert y-axis to show the highest count at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

#Creating a bubble chart for country distribution
plt.figure(figsize=(8, 6))
plt.scatter(country_counts.index[:10], country_counts.values[:10], s=country_counts.values[:10] * 10, alpha=0.6, color='blue', edgecolors='black')
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.title("Netflix Titles Distribution by Country (Bubble Chart)")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Storytelling Task:

#Insights of Horizontal Bar Chart:
#Clarity of Direct Comparisons: The bar chart excels in allowing viewers to make precise comparisons between countries. 
#It is easy to see which country has the most titles and how they stack up against each other. 
#This clarity is particularly beneficial when the differences in title counts are subtle, as viewers can quickly assess the lengths of the bars.
#Simplicity: The straightforward nature of the bar chart makes it accessible to a wide audience.
#Viewers do not need to interpret geographical data; they can focus solely on the numbers.

#Insights of Bubble Chart:
#Visual Impact: The bubble chart provides a visually engaging way to represent data. 
#The size of the bubbles immediately draws attention to countries with a large number of titles, making it easy to identify key players in the content landscape.
#Relative Comparison: While the bubble chart may not provide the same level of precision as the bar chart, it allows viewers to quickly gauge the relative size of different countries' title counts. 
#This can be particularly useful for identifying trends and patterns in content distribution.

#When comparing the two visualizations, the horizontal bar chart provides clearer insights into direct comparisons of title counts. 
#Its ability to display exact numbers makes it an effective tool for understanding which countries dominate in terms of content availability.
#On the other hand, the bubble chart excels at highlighting regional diversity and the relative scale of content distribution. 
# It visually emphasizes the differences in title counts, making it easier to spot trends and identify countries with significant content offerings.
#Hence, both visualizations serve important roles in analyzing the distribution of Netflix titles by country. 
#The horizontal bar chart is more effective for precise comparisons and understanding exact title counts, while the bubble chart provides a more engaging visual representation that highlights regional diversity and relative scale.
#For analysis of regional insights, I recommend using both visualizations in parallel.
#The bar chart can provide a clear overview of the top countries and their title counts, while the bubble chart can offer a broader perspective on the distribution of titles across different regions. 
# Together, they create a richer understanding of the content landscape on Netflix, guiding strategic decisions for content acquisition and marketing.
```

```python
#ANSWER 4: Word Cloud Analysis of Content Descriptions

#Dataset & Relevant Columns:

#Loading the Netflix dataset
netflix_data = pd.read_csv('netflix_titles.csv')

#Checking if the NLTK stopwords is downloaded
import nltk
nltk.download('stopwords')

#Separating Movies and TV Shows
movies = netflix_data[netflix_data['type'] == 'Movie']
tv_shows = netflix_data[netflix_data['type'] == 'TV Show']

#Function to preprocess text
def preprocess_text(text):
    # Removing punctuation and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Converting to lowercase
    text = text.lower()
    return text

#Preprocessing descriptions
movies['cleaned_description'] = movies['description'].apply(preprocess_text)
tv_shows['cleaned_description'] = tv_shows['description'].apply(preprocess_text)

#Combining all descriptions into a single string for each type
movies_text = ' '.join(movies['cleaned_description'])
tv_shows_text = ' '.join(tv_shows['cleaned_description'])

#Removing stop words
stop_words = set(stopwords.words('english'))
movies_text = ' '.join([word for word in movies_text.split() if word not in stop_words])
tv_shows_text = ' '.join([word for word in tv_shows_text.split() if word not in stop_words])

#Generating Word Clouds
plt.figure(figsize=(12, 6))

#Word Cloud for Movies
plt.subplot(1, 2, 1)
wordcloud_movies = WordCloud(width=800, height=400, background_color='white').generate(movies_text)
plt.imshow(wordcloud_movies, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Movies')

#Word Cloud for TV Shows
plt.subplot(1, 2, 2)
wordcloud_tv_shows = WordCloud(width=800, height=400, background_color='white').generate(tv_shows_text)
plt.imshow(wordcloud_tv_shows, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for TV Shows')

plt.tight_layout()
plt.show()
```

```python
#ASWER 5: Content Duration Analysis (Movies Only)

#Dataset & Relevant Columns:

#Loading the Netflix dataset
netflix_data = pd.read_csv('netflix_titles.csv')

#Filtering for Movies only
movies = netflix_data[netflix_data['type'] == 'Movie']

#Setting the duration and convering to numeric
movies = movies.dropna(subset=['duration'])
movies['duration'] = movies['duration'].str.replace(' min', '').astype(int)

# Analyzing the distribution of movie durations
plt.figure(figsize=(12, 6))
sns.histplot(movies['duration'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Movie Durations on Netflix')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
plt.figure(figsize=(12, 6))
sns.boxplot(x=movies['duration'], color='lightgreen')
plt.title('Boxplot of Movie Durations on Netflix')
plt.xlabel('Duration (minutes)')
plt.grid(axis='x', alpha=0.75)
plt.show()
```

```python
#Visualization Tasks:

#Creating a histogram to show the distribution of movie durations
plt.figure(figsize=(12, 6))
sns.histplot(movies['duration'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Movie Durations on Netflix')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

#Creating a box plot to illustrate the same data, emphasizing medians and outliers
plt.figure(figsize=(12, 6))
sns.boxplot(x=movies['duration'], color='lightgreen')
plt.title('Boxplot of Movie Durations on Netflix')
plt.xlabel('Duration (minutes)')
plt.grid(axis='x', alpha=0.75)
plt.show()
```

```python
#Storytelling Task:

#Histogram Analysis:
#The histogram presents a detailed view of the frequency distribution of movie durations. 
#Each bar represents a range of durations, allowing viewers to see how many movies fall within specific time intervals. 
#The addition of a kernel density estimate (KDE) line smooths the distribution, providing a clearer picture of the overall shape of the data.

#Insights from the Histogram:
#Central Tendency: The histogram allows for a visual assessment of the central tendency, showing where the bulk of movie durations lie. 
#Viewers can easily identify the most common duration ranges, which often cluster around a specific interval.
#Spread: The width of the histogram provides insights into the spread of durations. 
#By observing the height of the bars, viewers can gauge how many movies fall into shorter or longer duration categories.
#Trends: The histogram effectively highlights trends in movie durations, such as whether there is a preference for shorter or longer films. 
#It can reveal patterns, such as a peak in the number of movies around a certain duration, indicating a common industry standard.

#Box Plot Analysis:
#In contrast, the box plot offers a more concise summary of the data's central tendency and spread. It displays the median, quartiles, and potential outliers in a single visual. 
#The box represents the interquartile range (IQR), while the line inside the box indicates the median duration. Whiskers extend to show the range of the data, and any points outside this range are marked as outliers.

#Insights from the Box Plot:
#Central Tendency: The box plot clearly indicates the median duration, providing a quick reference for the central tendency of the data.
#Spread: The IQR displayed in the box plot highlights the spread of the middle 50% of the data, allowing viewers to understand how concentrated or dispersed the durations are.
#Outliers: The box plot excels at identifying outliers, which are easily visible as individual points beyond the whiskers. 
#This feature is particularly useful for spotting anomalies in movie durations that deviate significantly from the norm.

#When comparing the two charts, the box plot provides clearer insights into central tendency, spread, and outliers. 
#Its ability to summarize the data makes it an effective tool for quickly assessing the overall distribution of movie durations. The median and quartiles are easily identifiable, and the presence of outliers is immediately apparent.
#However, the histogram is more effective for identifying trends and patterns in the data. 
#It allows viewers to see the frequency of different duration ranges, making it easier to spot clusters and gaps in the data. The histogram's detailed representation of the distribution shape can reveal insights about audience preferences for movie lengths.
#Hence, both visualizations serve important roles in analyzing movie durations on Netflix. 
#The box plot is more effective for summarizing central tendency and identifying outliers, while the histogram excels at revealing trends and patterns in the data. For a comprehensive analysis, it may be beneficial to use both visualizations in tandem. 
#The box plot can provide a quick overview of the data's characteristics, while the histogram can offer deeper insights into the distribution and frequency of movie durations. Together, they create a richer understanding of the landscape of movie lengths available on the platform, guiding content creators and marketers in their strategies.
```
