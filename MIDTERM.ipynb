```python
#SECTION 2:
#TASK 1:

import pandas as pd
import numpy as np

# Loading the IMDb dataset
df = pd.read_csv('imdb_top_1000.csv')

# Displaying the number of rows and columns
num_rows, num_columns = df.shape
print('Number of rows: ', num_rows)
print('Number of columns: ', num_columns)

# Showing the first five rows of the dataset
print(df.head())

# Findng the top 5 most frequently occurring movie genres
genre_counts = df['Genre'].str.split(',').explode().str.strip().value_counts()
top_5_genres = genre_counts.head(5)
print('Top 5 most frequently occurring movie genres: ')
print(top_5_genres)
```

```python
#TASK 2:
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('imdb_top_1000.csv')

# Creating a histogram of IMDb ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['IMDB_Rating'], bins=30, kde=True)  
plt.title('Distribution of IMDb Ratings')
plt.xlabel('IMDb Ratings')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Describing the distribution
ratings_description = df['IMDB_Rating'].describe()
print("Summary statistics of IMDb ratings: ")
print(ratings_description)
skewness = df['IMDB_Rating'].skew()
print("Skewness of IMDb ratings: ", skewness)

#This histogram depicts the distribution of IMDb ratings, revealing that most films have ratings ranging from 7.5 to 8.5, with a peak at 7.8. 
#The KDE line shows a right-skewed distribution, which means that fewer films earn really high ratings.
```

```python
#TASK 3:
df = pd.read_csv('imdb_top_1000.csv')

#Creating a bar plot of the top 10 movie genres
genre_counts = df['Genre'].str.split(',').explode().str.strip().value_counts()
top_10_genres = genre_counts.head(10)
plt.figure(figsize=(10, 6))
top_10_genres.plot(kind='bar', color='skyblue')
plt.title('Top 10 Movie Genres')
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()

#This bar graph shows the first 10 most common cinematographic genres, in which the drama is more common with a large margin.
#Other popular genres are comedy, crime and adventure, while animation seems the least among the top 10.
```

```python
#TASK 4:
df = pd.read_csv('imdb_top_1000.csv')

#Creating a line plot showing the number of movies released each year
df['Released_Year'] = pd.to_numeric(df['Released_Year'], errors='coerce')
df = df.dropna(subset=['Released_Year'])
movies_per_year = df['Released_Year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o', linestyle='-', color='b')
plt.title('Number of Movies Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.grid()
plt.show()

#This line card shows the trend of film publications over time with a significant increase after the 1980s and reached a highlight in the 2000s. 
#The strong decline in the end has been due to incomplete data in recent years.
```

```python
#TASK 5:
df = pd.read_csv('imdb_top_1000.csv')

#Finding the top 10 directors with the most movies.
director_counts = df['Director'].value_counts()
top_10_directors = director_counts.head(10)
print('Top 10 directors: ', top_10_directors)

# Creating a bar plot to visualize the top 10 directors
plt.figure(figsize=(10, 6))
top_10_directors.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Directors with the Most Movies')
plt.xlabel('Directors')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.75)
plt.show()

#In this bar diagram, the number of films that are conducted by each of the top 10 directors is visualized, the bar height represents the count and director names along the horizontal axis. 
#It effectively compares the increase in these directors with a simple, evaluated representation of their filmographies.
```

```python
#TASK 6:
df = pd.read_csv('imdb_top_1000.csv')

#Creating a scatter plot to analyze the relationship
df = df.dropna(subset=['IMDB_Rating', 'No_of_Votes'])
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['IMDB_Rating'], y=df['No_of_Votes'], alpha=0.5)
sns.regplot(x=df['IMDB_Rating'], y=df['No_of_Votes'], scatter=False, color='red')
plt.title('IMDb Rating vs. Number of Votes')
plt.xlabel('IMDb Rating')
plt.ylabel('Number of Votes')
plt.xscale('log')  
plt.show()

#This scatter plot visualizes the relationship between IMDB assessments and the number of votes received by films, whereby each point represents a film and its position that indicates the respective values ​​on both axes. 
#A line of the best fit is covered to suggest a positive correlation between the two variables.
```

```python
#TASK 7:
df = pd.read_csv('imdb_top_1000.csv')

#Identifying the top 10 highest-grossing movies and visualizing them using a bar chart.
df['Gross'] = df['Gross'].astype(str).str.replace(',', '').astype(float)
df = df.dropna(subset=['Gross', 'Series_Title'])
top_10_grossing = df.sort_values(by='Gross', ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x='Gross', y='Series_Title', data=top_10_grossing, palette='viridis')
plt.title('Top 10 Highest-Grossing Movies', fontsize=14)
plt.xlabel('Gross Revenue (in $)', fontsize=12)
plt.ylabel('Movie Title', fontsize=12)
plt.show()

#This horizontal bar diagram attracts the 10 best films with the highest income according to your income, according to which beam lengths represent the size of the income and the gradation of colors, which indicates a relative position. 
#The clear title, the axis labels and the sorted beams quickly allow the classification list and the comparative financial success of each film.
```

```python
#TASK 8:
df = pd.read_csv('imdb_top_1000.csv')

#Creating a scatter plot of IMDb ratings vs. runtime and adding a trendline to analyze if longer movies tend to have higher ratings.
df['Runtime'] = df['Runtime'].str.replace(' min', '').astype(float) 
df['Runtime'] = pd.to_numeric(df['Runtime'], errors='coerce')
df = df.dropna(subset=['IMDB_Rating', 'Runtime'])
plt.figure(figsize=(10, 6))
sns.regplot(x=df['Runtime'], y=df['IMDB_Rating'], scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title('IMDb Rating vs. Runtime', fontsize=14)
plt.xlabel('Runtime (minutes)', fontsize=12)
plt.ylabel('IMDb Rating', fontsize=12)
plt.show()

#This scatter plot displays the relationship between IMDB notes and the runtime of film, each point representing the notation and the length of a film. 
#A linear regression line with a confidence interval band is superimposed to illustrate the general trend and the potential correlation between the two variables.
```
